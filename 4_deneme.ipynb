```python
import requests


url = 'https://www.isyatirim.com.tr/_Layouts/15/IsYatirim.Website/Common/ChartData.aspx/IndexHistoricalAll'


params = {
    'period': '1440',
    'from': '20220830000000',
    'to': '20230906235959',
}


instruments = ['XAU/USD', 'BRENT', 'XAG/USD', 'XPD/USD', 'XPT/USD']


data_by_instrument = {}


for instrument in instruments:

    params['endeks'] = instrument
    

    response = requests.get(url, params=params)
    

    if response.status_code == 200:

        data_by_instrument[instrument] = response.json()
    else:
        print(f"Failed to retrieve data for {instrument}. Status code: {response.status_code}")

# Now, data_by_instrument contains the data for each instrument in a dictionary format
# You can access and work with the data as needed for your analysis.
```

```python
# Verilen kodun √ßƒ±ktƒ±larƒ±nƒ± g√∂steren kod
for instrument, data in data_by_instrument.items():
    print(f"Data for {instrument}:")
    print(data)
    print("\n")
```

```python

# finans tarafƒ±ndak≈ü url im
url = 'https://www.isyatirim.com.tr/_Layouts/15/IsYatirim.Website/Common/ChartData.aspx/IndexHistoricalAll'

# ƒ∞stek i√ßin parametreleri tanƒ±mladƒ±m
params = {
    'period': '1440',
    'from': '20220830000000',
    'to': '20230906235959',
}

# veri √ßekmek istediklerim
instruments = ['XAU/USD', 'BRENT', 'XAG/USD', 'XPD/USD', 'XPT/USD']

# dataframeleri g√∂mmek i√ßin bo≈ü liste
dataframes = []

# ve d√∂nd√º 
for instrument in instruments:
    params['endeks'] = instrument
    
    # GET iel istek olu≈üturup veri al
    response = requests.get(url, params=params)
    
    # istek ba≈üarƒ±lƒ± mƒ±
    if response.status_code == 200:
        # webten gelen JSON ƒ± python dictionary e √ßevir-> verielri daha kolay i≈ülerim
        data = response.json()
        
        # ≈üimdi verileri dataframe e d√∂n√º≈üt√ºrmek i√ßin 
        df = pd.DataFrame(data['data'], columns=['timestamp', 'price'])
        
        #zaman kƒ±smƒ±ndaki sa√ßma sayƒ±larƒ± d√ºzeltme
        # Unix zaman damgasƒ± formatƒ±ndan, tarih zaman formatƒ±na √ßevirir
        df['timestamp'] = df['timestamp'].apply(lambda x: datetime.fromtimestamp(x / 1000).strftime('%Y-%m-%d %H:%M:%S'))
        
        #instrument adlƒ± bi satur ekledim
        df['instrument'] = instrument
        
       
        dataframes.append(df)
    else:
        print(f"Failed to retrieve data for {instrument}. Status code: {response.status_code}")

# birden fazla tablo birle≈ütirme
combined_df = pd.concat(dataframes, ignore_index=True)

print(combined_df)
```

```python
import requests
import pandas as pd
from datetime import datetime

url = 'https://www.isyatirim.com.tr/_Layouts/15/IsYatirim.Website/Common/ChartData.aspx/IndexHistoricalAll'


params = {
    'period': '1440',
    'from': '20220830000000',
    'to': '20230906235959',
}


instruments = ['XAU/USD', 'BRENT', 'XAG/USD', 'XPD/USD', 'XPT/USD']

# Initialize an empty list to store dataframes
dataframes = []


for instrument in instruments:

    params['endeks'] = instrument
    

    response = requests.get(url, params=params)
    

    if response.status_code == 200:
       
        data = response.json()
        

        df = pd.DataFrame(data['data'], columns=['timestamp', f'{instrument}'])
        

        df['timestamp'] = df['timestamp'].apply(lambda x: datetime.fromtimestamp(x / 1000).strftime('%Y-%m-%d'))

        # df['instrument'] = instrument
        

        dataframes.append(df)
        
    else:
        print(f"Failed to retrieve data for {instrument}. Status code: {response.status_code}")

# combined_df yi yatay olrak birle≈ütirmete √ßalƒ±≈ü
combined_df = pd.concat(dataframes, axis=1)
##########################
# instrument adlarƒ±nƒ± i√ßerecek ≈üekilde yeniden 
#combined_df.columns = [f'{col[0]}_{col[1]}' for col in combined_df.columns]

#######################
print(combined_df)
```
